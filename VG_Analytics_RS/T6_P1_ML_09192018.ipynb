###Analysis and Testing

#Chi-squared test on Genre

```python
#Let's first create a new variable "Other Countries" which is Global Sales minus US Sales
sales_by_genre['OtherCountries'] = sales_by_genre['Global_Sales'] - sales_by_genre['NA_Sales']
sales_by_name['OtherCountries'] = sales_by_name['Global_Sales'] - sales_by_name['NA_Sales']
sales_by_platform['OtherCountries'] = sales_by_platform['Global_Sales'] - sales_by_platform['NA_Sales']
sales_by_year['OtherCountries'] = sales_by_year['Global_Sales'] - sales_by_year['NA_Sales']

#Let's grab the row count using shape and taking the row dimension as this will be an input into our Chi Squared Test
Rows_Genre = sales_by_genre.shape[0]
Rows_Name = sales_by_name.shape[0]
Rows_Platform = sales_by_platform.shape[0]
Rows_Year = sales_by_year.shape[0]
```

Import dependency (this could have been done at the beginning as well) we would like to utilize chi squared to look at differences in sales between populations

```python

import scipy.stats as stats
critical_value_genre = round(stats.chi2.ppf(q = 0.95, df = Rows_Genre -1 ),0)
critical_value_name = round(stats.chi2.ppf(q = 0.95, df = Rows_Name-1),0)
critical_value_Platform = round(stats.chi2.ppf(q = 0.95, df = Rows_Platform-1),0)
critical_value_Year = round(stats.chi2.ppf(q = 0.95, df = Rows_Year-1),0)

print(f'Genre DF = {critical_value_genre}; \
Value DF = {critical_value_name},\
Platform DF = {critical_value_Platform},\
Year DF = {critical_value_Year}')

#Notice how the degrees of freedom go up with larger datasets. 
```

#Running chi squared test on each of the four sub data sets to see if there is a difference between NA Sales and All Other
#Countries by Genre, Name, Platform and Year

```python
Genre_CS = stats.chisquare(sales_by_genre['OtherCountries'], sales_by_genre['NA_Sales'])
Name_CS = stats.chisquare(sales_by_name['OtherCountries'], sales_by_name['NA_Sales'])
PLatform_CS = stats.chisquare(sales_by_platform['OtherCountries'], sales_by_platform['NA_Sales'])
Year_CS = stats.chisquare(sales_by_year['OtherCountries'], sales_by_year['NA_Sales'])
```

Influence plots
Influence plots show the (externally) studentized residuals vs. the leverage of each observation as measured by the hat matrix.
 
The influence of each point can be visualized by the criterion keyword argument. Options are Cook's distance and DFFITS, two measures of influence.
